## 6. 데이터 프레임 병합
    실제 분석업무를 진행하다보면 데이터가 여기저기 분산되어 있는 경우가 더 많습니다.
    조각난 데이터를 분석에 필요한 dataset으로 만들기 위해 dataframe 병합을 많이 사용합니다.
    한 개 이상의 dataframe을 병합할 때 주로 사용하는 함수 2가지를 알아보겠습니다.

### 6-1. 데이터 병합에 사용 가능한 key(병합할 기준이 되는 행 or 열) 값이 있는 경우
    pd.merge(base dataframe, merge할 dataframe)
    
    사용 가능한 parameter
    - how : 'left', 'right', 'inner', 'outer'
    - left_on : key 값이 다를 경우 base dataframe의 key 설정
    - right_on : key 값이 다를 경우 merger할 datafraem의 key 설정

```python
# 두 개의 merger할 dataframe을 생성합니다.
merge_df1 = pd.DataFrame({
    '이름' : ['원영', '사쿠라', '유라', '예나', '유진', '나코', '은비', '혜원', '히토미', '채원', '민주', '째욘'],
    '국어': [100, 70, 70, 70, 60, 90, 90, 70, 70, 80, 100, 100],
    '영어': [100, 90, 80, 50, 70, 100, 70, 90, 100, 100, 80, 100]
    }, columns = ['이름' ,'국어', '영어'])

merge_df2 = pd.DataFrame({
    '일어': [80, 100, 100, 90, 70, 50, 100],
    '수학': [90, 70, 100, 80, 70, 80, 90],
    '이름': ['원영', '사쿠라', '나코', '히토미', '예나', '은비', '째욘'],
    }, columns=['일어', '수학', '이름'])

pd.merge(merge_df1, merge_df2)
```

### 8-1. column 삭제
    현재 dataset에는 개인식별 정보가 지워져서 데이터가 존재하지 않습니다. 불필요한 데이터 columns를 지워보도록 하겠습니다.

```python
person_df.info()
```

```python
# 지울 column의data값이 모두 NaN인지 확인한다.
# 제가 사용하는 dataset은 예시와 다르기 때문에 모두 data들이 가득차있습니다.
# person_df[''].sum(), person_df[''].sum ...
```

```python
# column 삭제 drop('column name', axis=1)
# del(df['coluimn name'])
# 실제로는 column 및 row도 삭제 가능합니다. axis=0 (Default)
# inplace=True parameter를 사용해서 원본값을 변경 가능합니다.
person_df = person_df.drop('Lot', axis=1)
# person_df = person_df.drop('id', axis=1)
```

```python
person_df
# Lot column이 삭제된 결과를 확인 가능합니다.
```

```python
# inplace parameter 사용하여 원본 데이터에도 반영 가능하다.
person_df.drop('AM or PM', axis=1, inplace=True)
```

```python
# del memory 삭제 키워드 사용
del person_df['Browser Info']
del person_df['Company']
```

```python
person_df
```

### 8-2. Column name 변경
